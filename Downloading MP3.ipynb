```python
from pytube import YouTube
from youtube_search import YoutubeSearch
import argparse
import pydub
from pydub import AudioSegment
import json
import ffmpeg 
import pandas as pd
import csv
import os
```

```python
dataset = pd.read_csv('hello.csv', encoding = "ISO-8859-1")
dataset
```

```python
for x in range(31, 94):
    id = dataset.iloc[x,0]
    name = dataset.iloc[x,1]
    artist = dataset.iloc[x,2]
    print(name, artist)
    
    fullname = name + " "+ artist
    #searching the song on youtube and getting the url
    results = YoutubeSearch(fullname, max_results = 1).to_json()
    json_string = json.loads(results)
    ans = json_string["videos"][0]["url_suffix"]
    url = "youtube.com" + ans
    print(url)

    #downloading mp4 song from youtube url
    yt = YouTube(url)
    stream = yt.streams.get_by_itag(18)
    print('Download started. Wait... '+str(x))
    filename = stream.download()
    file_name = filename[33:]
    files_path = 'mp3music'
    startMin = 0
    startSec = 5

    endMin = 0
    endSec = 50
    #for file_name in files_path:
            # Time to miliseconds
    startTime = startMin*60*1000+startSec*1000
    endTime = endMin*60*1000+endSec*1000
    
    # Opening file and extracting segment
    print (file_name)
    song = AudioSegment.from_file(file_name)
    extract = song[startTime:endTime]

    # Saving
    new_name = str(id) + '.mp3'
    extract.export(new_name, format="mp3")
    os.remove(file_name)

print ("done")
```
